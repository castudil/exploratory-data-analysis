```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

# Generate 100 random points from a uniform distribution between 0 and 10
np.random.seed(42)  # Set a random seed for reproducibility
x = np.random.uniform(0, 10, size=100)

# Print the first 10 points
print(x[:10])
```

In this program, we use the numpy.random.uniform function to generate 100 random numbers from a uniform distribution between 0 and 10. The size parameter specifies the size of the array we want to generate, which in this case is 100.

We also set a random seed using the np.random.seed function to ensure that the random numbers are reproducible.

Finally, we print the first 10 numbers using array slicing to verify that the array was generated correctly. When you run this program, it will print the first 10 random numbers to the console.

```python
# Generate y points following the function y = 3*x + 1 with added noise
y = 3*x + 1 + np.random.normal(0, 3, size=100)

# Print the first 10 y points
print(y[:10])
```

We then generate the y values using the same function as before, but with an added noise term generated by np.random.normal. We pass in the mean and standard deviation of the normal distribution we want to sample from, as well as the size of the array of noise values we want to generate.

Finally, we create a scatter plot of the x and y values with added noise using Matplotlib, and display the plot using the show function. When you run this program, it will display a scatter plot of the x and y values with added noise.

```python
# Plot x and y
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

We create a scatter plot of the x and y values using the scatter function from Matplotlib. We also add x and y axis labels using the xlabel and ylabel functions.

Also, we show the plot using the show function. When you run this program, it will display a scatter plot of the x and y values.

```python
# Train a linear regression model
reg = LinearRegression().fit(x.reshape(-1, 1), y)

# Print the coefficient(s) and intercept
print("Coefficient(s): ", reg.coef_)
print("Intercept: ", reg.intercept_)
```

We then train a linear regression model using the LinearRegression class from scikit-learn. We pass in the x values as a 2D array with one column using x.reshape(-1, 1) to fit the model.

Finally, we print the coefficient(s) and intercept of the trained model using the coef_ and intercept_ attributes of the LinearRegression object. When you run this program, it will print the coefficient(s) and intercept to the console.

```python
# Plot the scatter plot of x and y with added noise
plt.scatter(x, y)
# Plot the line of best fit predicted by the trained linear regression model
x_test = np.linspace(0, 10, 100)  # Generate 100 test points to plot the line
y_pred = reg.predict(x_test.reshape(-1, 1))  # Predict the y values for the test points
plt.plot(x_test, y_pred, color='red')

# Add axis labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')

# Show the plot
plt.show()
```
